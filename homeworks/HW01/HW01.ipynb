```python
import numpy as np #Подключаем библиотеку


#Создаём различные массивы
TwoD_normal_array = np.array([[0, 1, 2], [3, 4, 5]])
print("Двумерный массив из обычного списка:\n", TwoD_normal_array)

zero_array = np.zeros(3)
print("Массив из нулей:", zero_array)

one_array = np.ones(3)
print("Массив из едениц:", one_array)

arange_array = np.arange(0, 10, 2)
print("Массив созданный arange:", arange_array)
print("------------------------")


#Демонстрируем атрибуты
print("Наш массив:\n", TwoD_normal_array)
print("Форма массива:", TwoD_normal_array.shape)
print("Тип данных:", TwoD_normal_array.dtype)
print("------------------------")


#Примеры индексирования и срезов
print("Наш массив:\n", TwoD_normal_array)
print("Отдельный элемент из 2 строки и 3 столбца:", TwoD_normal_array[1, 2])
print("Извлечённый подмассив:\n", TwoD_normal_array[0:2, 0:2])
print("------------------------")


#Примеры векторных операций и базовых статистик
first_array = np.array([1, 2, 3, 4, 5])
second_array = np.array([10, 20, 30, 40, 50])
print("Наши массивы:\n", first_array, "\n", second_array)

summa = first_array + second_array
print("Сложение наших массивов:", summa)

multiplication = first_array * second_array
print("Произведение наших массивов:", multiplication)

#Базовые статистики первого массива
print("\nНаш массив:", first_array)
print("Среднее значение:", np.mean(first_array))
print("Минимальное значение:", np.min(first_array))
print("Сумма всех элементов:", np.sum(first_array))
print("Медиана:", np.median(first_array))
```
